Research Interests and Questions

In 2011, a drought recorded as the worst in 60 years brought a food crisis to Africa. A long-standing conflict between two East African ethnic groups, the Turkana and the Pokot, intensified during the drought. Both groups rely on the cattle they herd to survive. The drought led to desertification in Kenyan cattle grazing land, pushing Turkana cattle herders closer to Pokot territory. The Pokot organized raids on Turkana cattle herders in Kenya and drove the cattle across the border to Uganda.
This conflict raises questions about the relationships between Kenya, Uganda, and the other countries of East Africa. What tone did the interactions between Uganda and Kenya take during the drought, and how did other countries in the region react to the conflict?


Hypothesis

Because a severe drought struck East Africa in 2011, conflicts directed toward Kenya by other countries in the region intensified during 2011. Both verbal and material conflicts became more common and more intense.


GDELT, a database of major world events, stores all of its data on world events with Google BigQuery. The SQL query below searches GDELT's BigQuery database for verbal and material conflict events directed toward Kenya in 2011. It also includes mapping information on the country which incited the event.

```python
from google.cloud import bigquery

client = bigquery.Client()
```

```python
sql = """
SELECT
    Year,
    Actor1Code,
    Actor2Code,
    QuadClass,
    Actor1Geo_Lat,
    Actor1Geo_Long
FROM
    `gdelt-bq.full.events`
WHERE
    Actor2Code = 'KEN'
    AND Year = 2011
    AND (QuadClass = 3 OR QuadClass = 4)

"""
df = client.query(sql).to_dataframe()
```

Show the first few rows of data.

```python
# The table contains some bad data at latitude, longitude (0, 0) that should be removed
df = df.query('Actor1Geo_Lat != 0 and Actor1Geo_Long != 0')

df.head(9)
```

Use pandas to aggregate the data by location. Count the numbers of material and verbal events that started at the same latitude and longitude.

```python
import pandas as pd

pivot = pd.pivot_table(df, index=['Actor1Geo_Lat', 'Actor1Geo_Long'], columns='QuadClass', aggfunc='size')
pivot = pivot.fillna(0)
pivot = pivot.reset_index()

pivot.head(9)
```

Make the data more usable for mapping. Sort the data to find which 100 locations started the most conflicts and label the conflict columns.

```python
top_100_Verbal = pivot.sort_values(by=[3], ascending=False)[0:101]
top_100_Material = pivot.sort_values(by=[4], ascending=False)[0:101]

top_100_Verbal = top_100_Verbal.rename(columns={3: 'Verbal_Conflicts',
                                                4: 'Material_Conflicts'})
top_100_Material = top_100_Material.rename(columns={3: 'Verbal_Conflicts',
                                                4: 'Material_Conflicts'})
```

As an example of what the data looks like before any mapping, show some rows sorted by the number of verbal conflict events.

```python
top_100_Verbal[:10]
```

Plot some initial data on a map of Africa.

```python
import cartopy
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

plt.figure(figsize=(15, 15))


#     Create a map using the Plate Caree map projection.
# Map projections transform the curved surface of the Earth into a flat map using latitude and longitude.
# Plate Carrée is a map projection commonly used for satellite imagery.
ax = plt.axes(projection=cartopy.crs.PlateCarree()) 


#     Add geographic details to the map.
ax.add_feature(cartopy.feature.LAND)
ax.add_feature(cartopy.feature.COASTLINE)
ax.add_feature(cartopy.feature.BORDERS, linestyle='dotted')
ax.add_feature(cartopy.feature.OCEAN)   # Bodies of water may have been points of conflict during the drought
ax.add_feature(cartopy.feature.LAKES)
ax.add_feature(cartopy.feature.RIVERS)


#     Plot locations on the map that took part in conflicts directed toward Kenya.
# The size of each point is calculated using the frequency of conflicts in the area.


for i in range(99):
    plt.plot(top_100_Material['Actor1Geo_Long'].values[i],
              top_100_Material['Actor1Geo_Lat'].values[i],
              'or',
              markersize=top_100_Material['Material_Conflicts'].values[i] / 20,
              alpha=.3)
    plt.plot(top_100_Verbal['Actor1Geo_Long'].values[i],
              top_100_Verbal['Actor1Geo_Lat'].values[i],
              'oy',
              markersize=top_100_Verbal['Verbal_Conflicts'].values[i] / 20,
              alpha=.8)

ax.set_extent([-20, 60, -30, 45])


# Add descriptive elements to the map to show where it is oriented in space

# give the map a title
ax.set_title("Sources of Conflict Directed Toward Kenya in 2011", fontsize=24)

# make a legend with labels for the map elements
yellow_circle = mpatches.Circle(color='yellow', alpha=.8, xy=(0,0), radius=1, label='Verbal Conflicts')
red_circle = mpatches.Circle(color='red', alpha=.3, xy=(0,0), radius=1, label='Material Conflicts')
ax.legend(handles=[yellow_circle, red_circle], loc='upper left', fontsize=16)

# cite GDELT as the data source
plt.text(0, 0, 'Data Source: GDELT', fontsize=16, transform=ax.transAxes, horizontalalignment='left',
         verticalalignment='bottom', bbox=dict(boxstyle="Square,pad=0.3", fc='white', ec='black'));

# Add names to the African countries
shpfilename = cartopy.io.shapereader.natural_earth(resolution='110m',
                                                category='cultural',
                                                name='admin_0_countries')
reader = cartopy.io.shapereader.Reader(shpfilename)
countries = reader.records()

for country in countries:
    if country.attributes['CONTINENT'] == 'Africa':
        x = country.geometry.centroid.x
        y = country.geometry.centroid.y
        if country.attributes['NAME'] == 'Dem. Rep. Congo' or country.attributes['NAME'] == 'South Africa':
            ax.text(x, y, country.attributes['NAME'], color='black', size=11, ha='center', va='center',
                transform=cartopy.crs.PlateCarree(), rotation=30)
        else:
            ax.text(x, y, country.attributes['NAME'], color='black', size=11, ha='center', va='center',
                transform=cartopy.crs.PlateCarree())
```

Observations

Conflict was the most frequent at the capital of Kenya, Nairobi. Kenya's internal conflicts were about equally verbal and material in nature. Other major conflicts took place in central Kenya and along the eastern border to Somalia. Several smaller conflicts are scattered along the Kenyan borders to Uganda and Tanzania, which are close to Pokot territory.

Outside of Kenya, the East African countries bordering Kenya were significant sources of material conflicts directed toward Kenya. Somalia confronted Kenya from many different cities and towns. Ethiopia, Uganda, and Tanzania also fought against Kenya.

Another origin of conflict near Khartoum, the capital of Sudan, is in an environment especially vulnerable to drought.

Analysis

All of the countries that frequently started conflicts with Kenya have two traits in common: close proximity to Kenya, and vulnerability to drought.

The short distance between Kenya and its aggressors could have made it a target during the drought. Kenya was likely competing for resources with the other East African countries, especially Somalia, which started many material conflicts. Kenya and Somalia have fought since colonial times, and in 2011, Kenya invaded Somalia (1). The drought must have contributed to the ongoing conflict between Kenya and Somalia.

The East African climate is a mix of tropical and desert-like conditions. Kenya, Somalia, and eastern Ethiopia are especially arid (2). Sudan is part of the Sahel region, which has a sensitive, transitional climate between the Sahara desert to the north and a savanna to the south. Tropical countries didn't participate in conflicts with Kenya nearly as much as arid countries. Tanzania and Uganda in particular both have tropical, savannah climates and no deserts (2).

The drought was in all likelihood a major factor in all of the conflicts that took place in East Africa during 2011, but it is difficult to measure its direct impact on the region.
